```python
import os
import tweepy
import json
import credentials
```

```python
consumer_key= credentials.cons_key
consumer_secret= credentials.cons_secret
access_token= credentials.auth_token
access_token_secret= credentials.auth_secret
```

```python
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)
```

```python
search_terms = ['AMC', 'GME', 'CLOV', 'BB', 'CLF', 'CLNE', 'WKHS', 'SENS', 'PLTR', 'RKT', 'WWE', 'FORD']
```

```python
#function modified from Michelle Aniuchi blogpost
#url = https://aadaobi.medium.com/exploring-twitter-api-and-data-using-tweepy-pandas-and-matplotlib-part-1-2ac07fcc4717

date_since = "2021-01-01"
# search_terms = ['doge']

def stream_tweets(search_term):
    data = [] # empty list to which tweet_details obj will be added
    counter = 0 # counter to keep track of each iteration
    for tweet in tweepy.Cursor(api.search, q='\"{}\" -filter:retweets'.format(search_term), count=100, 
                               lang='en', tweet_mode='extended', since=date_since).items():
        tweet_details = {}
        tweet_details['name'] = tweet.user.screen_name
        tweet_details['tweet'] = tweet.full_text
        tweet_details['retweets'] = tweet.retweet_count
        tweet_details['id'] = tweet.user.id
        tweet_details['created'] = tweet.created_at.strftime("%m-%d-%Y, %H:%M:%S")
        data.append(tweet_details)
        
        counter += 1
        if counter == 100000:
            break
        else:
            pass
    with open('data/{}.json'.format(search_term), 'w') as f:
        json.dump(data, f)
    print(f'{search_term} done!')
```

```python
if __name__ == "__main__":
    print('Starting to stream...')
    for search_term in search_terms:
        stream_tweets(search_term)
    print('finished!')
```
